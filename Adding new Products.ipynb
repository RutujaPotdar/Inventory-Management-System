## Adding new items in data

```python
# importing json file
import json
```

```python
# Open file in json formate
fd = open("record.json","r" )
r = fd.read()
fd.close()
```

```python
record = json.loads(r)
```

```python
# Take Product ID from user
prod_id = input("Enter Product ID : ")

# Add more Items in already existing items
if prod_id in record:
#     Product name
    print("Product name :" , record[prod_id]['name'])
    
#     Product Chategory
    if 'ctg' in record[prod_id] :
        print("Product Chategory : ",record[prod_id]['ctg'] )
    else:
        ctg = input("Enter Catagery : ")
#     Product Price    
    print("Product Price :" , record[prod_id]['pr'])
    
#     Add quantity
    qn = input("Enter Quantity : ")
    record[prod_id]['qn'] = int(record[prod_id]['qn']) + int(qn)
    print("Product Quantity : ",record[prod_id]['qn'] )
    
#     Discount
    if 'dis' in record[prod_id]:
        print("Product Discount : ",record[prod_id]['dis'] )
    else:
        dis = int(input("Enter Discount : "))
        
#     Expiry Date    
    exp_date = input("Enter Expiry Date : ")
    
#     Save record
    record[prod_id] = {'name': record[prod_id]['name'], 'ctg': record[prod_id]['ctg'] ,'pr': record[prod_id]['ctg'], 'qn': record[prod_id]['qn'],'dis': record[prod_id]['dis'],'exp_date':exp_date}

# Add new items in data    
else:
    name = input("Enter Name : ")
    ctg = input("Enter Catagery : ")
    pr = int(input("Enter Price  : "))
    qn = int(input("Enter Quantity : "))
    dis = int(input("Enter Discount : "))
    exp_date = input("Enter Expirt Date : ")
    record[prod_id] = {'name': name, 'ctg': ctg ,'pr': pr, 'qn': qn,'dis': dis,'exp_date':exp_date}

# Save data in given file
fl = json.dumps(record)
fd = open("record.json","w")
r = fd.write(fl)
fd.close()
```
