##  Sell items to Costumers

```python
# Import json file and read data existing in given file
import json

fd = open("record.json","r" )
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

```python
print("Hello ! , Costumer ")

print("------------------------")

prod_id  = input("Enter the product_Id: ")
ui_quant = int(input("Enter the quantity: "))

# if sufficient quantity is not available
if record[prod_id ]['qn']<ui_quant :
    print("Sorry, sufficiant quantity is not available")
else:
    record[prod_id ]['qn'] = int(record[prod_id ]['qn']) - ui_quant

print("***********************")

print("Product: ", record[prod_id]['name'])
print("Price: ", record[prod_id]['pr'])
print("Discount :" ,record[prod_id]['dis'] ,"%")
print("------------------------")

# Billing Amount with Discount
print("Billing Amount: ", (int(record[prod_id ]['pr'])-int(record[prod_id]['dis'])/100) * ui_quant)

print("***********************")
print("Thank You :)")

```

```python
# Reload file with changes
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
